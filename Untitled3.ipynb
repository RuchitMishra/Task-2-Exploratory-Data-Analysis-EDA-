```python
#import all required libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('Titanic-Dataset (2).csv')
print(df.head())
```

```python
print(df.dtypes)

print(df.isnull().sum())
```

```python
#working upon the null values

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print(df.isnull().sum())
```

```python
#since , for summary(mean, median ,mode ) , some columns ae not required. Therefore we have to remove them.

df.drop('PassengerId' , axis=1 , inplace=True)

# the other not required columns will be automatically excluded by the
# df.describe function while discribing.

print(df.head())
```

```python
#Generating summary statics.

summary = df.describe()
print(summary)
```

```python
#printing the ttoal no. of people who survived
survived = df[df['Survived']==1].shape[0]  #-----Only this syntax is giving the no. of survivers.
# [0] means row and [1] means column

#print(df['Survived'].value_counts().sum())   --- This line will not give the survivers , it will give the total no. of people.
#print(df['Survived'].nunique())   ------- This will also not give the no. of survivers.
print('Total no. of people who survived =', survived)
```

```python
#2.Creating histograms and boxplots for numeric features.

#2.1 Creating boxplots for numeric features.
plt.figure(figsize = (10,6))
sns.boxplot(x = df['Survived'])
plt.title("Box plot of Survived")
plt.show()

plt.figure(figsize = (10,6))
sns.boxplot(x = df['Age'])
plt.title("Box plot of Age")
plt.show()


plt.figure(figsize = (10,6))
sns.boxplot(x = df['Pclass'])
plt.title("Box plot of Pclass")
plt.show()

plt.figure(figsize = (10,6))
sns.boxplot(x = df['SibSp'])
plt.title("Box plot of SibSp")
plt.show()

plt.figure(figsize = (10,6))
sns.boxplot(x = df['Parch'])
plt.title("Box plot of Parch")
plt.show()

plt.figure(figsize = (10,6))
sns.boxplot(x = df['Fare'])
plt.title("Box plot of Fare")
plt.show()



```

```python
#2.Creating histograms and boxplots for numeric features.

#2.2 Creating histograms for numeric features.

plt.figure(figsize = (10,6))
sns.histplot(x = df['Survived'])
plt.title("Histogram of Survived")
plt.show()


plt.figure(figsize = (10,6))
sns.histplot(x = df['Age'])
plt.title("Histogram plot of Age")
plt.show()


plt.figure(figsize = (10,6))
sns.histplot(x = df['Pclass'])
plt.title("Histogram plot of Pclass")
plt.show()

plt.figure(figsize = (10,6))
sns.histplot(x = df['SibSp'])
plt.title("Histogram plot of SibSp")
plt.show()

plt.figure(figsize = (10,6))
sns.histplot(x = df['Parch'])
plt.title("Histogram plot of Parch")
plt.show()

plt.figure(figsize = (10,6))
sns.histplot(x = df['Fare'])
plt.title("Histogram plot of Fare")
plt.show()

```

```python
#4.Identify patterns, trends, or anomalies in the data.
#5.Make basic feature-level inferences from visuals.

import pandas as pd
df = pd.read_csv('Titanic-Dataset (2).csv')
Corr = df.corr(numeric_only=True)
print(Corr)
```

Strong negative correlaton between Sx(Male) and Survived.
Pclass, Fare and Sex ar top predictors of survival.

```python
df['Survived'].value_counts(normalize=True)
```

About 38% survied and 62% didn't .


```python
df.groupby('Sex')['Survived'].mean()
```

Females had a much heigher survival rate(74%) than males(19%). This means that women and the chldren were the first to be evacuated from the ship.

```python
df.groupby('Pclass')['Survived'].mean()
```

1st class passangers had the highest survival rate, then 2nd and then 3rd class. This means that rich were first favoured while evacuating.

```python
df['Family Size'] = df['SibSp'] + df['Parch']
df.groupby('Family Size')['Survived'].mean()
```

Passengers with small families(2-3) members had higher survival chancces tha those alone or with large families.
